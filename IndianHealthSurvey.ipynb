**INTERACTIVE DIVERGING BARPLOT:**Difference between urban and rural population,treatment,household,sexratio by state?

```python
import pandas as pd
import plotly.graph_objects as go

# Load your dataset
# Assuming your data is in a CSV file
data = pd.read_csv('/content/Key_indicator_districtwise.csv')

# Remove duplicate columns
df = data.loc[:, ~data.columns.duplicated()]
df = df.loc[:, ~df.columns.duplicated()]

# Calculate the differences between rural and urban for some columns
df['AA_Sample_Units_Difference'] = df['AA_Sample_Units_Rural'] - df['AA_Sample_Units_Urban']
df['AA_Population_Difference'] = df['AA_Population_Rural'] - df['AA_Population_Urban']
df['BB_Average_Household_Size_Difference'] = df['BB_Average_Household_Size_All_Rural'] - df['BB_Average_Household_Size_All_Urban']
df['CC_Sex_Ratio_Difference'] = df['CC_Sex_Ratio_At_Birth_Rural'] - df['CC_Sex_Ratio_At_Birth_Urban']
df['DD_Person_Difference'] = df['DD_Person_Rural'] - df['DD_Person_Urban']

# Select a subset of columns for the plot
columns_to_plot = [
    'AA_Sample_Units_Difference', 'AA_Population_Difference',
    'BB_Average_Household_Size_Difference', 'CC_Sex_Ratio_Difference', 'DD_Person_Difference'
]

# Normalize the data for coloring
df['AA_Sample_Units_Difference_norm'] = (df['AA_Sample_Units_Difference'] - df['AA_Sample_Units_Difference'].min()) / (df['AA_Sample_Units_Difference'].max() - df['AA_Sample_Units_Difference'].min())
df['AA_Population_Difference_norm'] = (df['AA_Population_Difference'] - df['AA_Population_Difference'].min()) / (df['AA_Population_Difference'].max() - df['AA_Population_Difference'].min())
df['BB_Average_Household_Size_Difference_norm'] = (df['BB_Average_Household_Size_Difference'] - df['BB_Average_Household_Size_Difference'].min()) / (df['BB_Average_Household_Size_Difference'].max() - df['BB_Average_Household_Size_Difference'].min())
df['CC_Sex_Ratio_Difference_norm'] = (df['CC_Sex_Ratio_Difference'] - df['CC_Sex_Ratio_Difference'].min()) / (df['CC_Sex_Ratio_Difference'].max() - df['CC_Sex_Ratio_Difference'].min())
df['DD_Person_Difference_norm'] = (df['DD_Person_Difference'] - df['DD_Person_Difference'].min()) / (df['DD_Person_Difference'].max() - df['DD_Person_Difference'].min())

# Create the diverging bar plot
fig = go.Figure()

color_scale = 'Turbo'  # Diverging color scale

for column in columns_to_plot:
    norm_column = column + '_norm'
    fig.add_trace(go.Bar(
        x=df['State_Name'],
        y=df[column],
        name=column,
        orientation='v',
        marker=dict(color=df[column], colorscale=color_scale, showscale=True, cmin=-1, cmax=1),
        hovertemplate='%{y}'
    ))

# Update layout for better readability
fig.update_layout(
    title='Diverging Bar Plot',
    xaxis_title='State Name',
    yaxis_title='Difference (Rural - Urban)',
    barmode='relative',
    height=800,
    width=1200,
    margin=dict(l=50, r=50, t=50, b=50),
)

# Show the plot
fig.show()

```

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from statsmodels.tsa.stattools import acf, pacf

# Sample data
np.random.seed(42)
dates = pd.date_range('2011-01-01', periods=100)
data = {
    'Date': dates,
    'KK_Having_Any_Kind_Of_Symptoms_Of_Chronic_Illness_Per_100000_Population_Person_Rural': np.random.randint(5, 20, size=100),
    'KK_Having_Any_Kind_Of_Symptoms_Of_Chronic_Illness_Per_100000_Population_Person_Urban': np.random.randint(3, 10, size=100),
    'KK_Having_Any_Kind_Of_Symptoms_Of_Chronic_Illness_Per_100000_Population_Male_Total': np.random.randint(40, 70, size=100),
    'KK_Having_Any_Kind_Of_Symptoms_Of_Chronic_Illness_Per_100000_Population_Male_Rural': np.random.randint(20, 50, size=100),
    'KK_Having_Any_Kind_Of_Symptoms_Of_Chronic_Illness_Per_100000_Population_Male_Urban': np.random.randint(10, 30, size=100),
    'KK_Having_Any_Kind_Of_Symptoms_Of_Chronic_Illness_Per_100000_Population_Female_Total': np.random.randint(40, 70, size=100),
    'KK_Having_Any_Kind_Of_Symptoms_Of_Chronic_Illness_Per_100000_Population_Female_Rural': np.random.randint(20, 50, size=100),
    'KK_Having_Any_Kind_Of_Symptoms_Of_Chronic_Illness_Per_100000_Population_Female_Urban': np.random.randint(10, 30, size=100),
    'KK_Having_Any_Kind_Of_Symptoms_Of_Chronic_Illness_And_Sought_Medical_Care_Person_Total': np.random.randint(40, 70, size=100),
    'KK_Having_Any_Kind_Of_Symptoms_Of_Chronic_Illness_And_Sought_Medical_Care_Person_Rural': np.random.randint(20, 50, size=100),
    'KK_Having_Any_Kind_Of_Symptoms_Of_Chronic_Illness_And_Sought_Medical_Care_Person_Urban': np.random.randint(10, 30, size=100),
    'KK_Having_Any_Kind_Of_Symptoms_Of_Chronic_Illness_And_Sought_Medical_Care_Female_Total': np.random.randint(40, 70, size=100),
    'KK_Having_Any_Kind_Of_Symptoms_Of_Chronic_Illness_And_Sought_Medical_Care_Male_Total': np.random.randint(40, 70, size=100),
}

df = pd.DataFrame(data)

# Choose a column to plot
column_to_plot = 'KK_Having_Any_Kind_Of_Symptoms_Of_Chronic_Illness_Per_100000_Population_Person_Rural'

# Calculate ACF and PACF values
lag_acf = acf(df[column_to_plot], nlags=20)
lag_pacf = pacf(df[column_to_plot], nlags=20)

# ACF Plot
fig_acf = go.Figure()
fig_acf.add_trace(go.Bar(
    x=np.arange(len(lag_acf)),
    y=lag_acf,
    marker_color='blue'
))

fig_acf.update_layout(
    title='Autocorrelation Function (ACF)',
    xaxis_title='Lag',
    yaxis_title='ACF',
    template='plotly_white'
)

# PACF Plot
fig_pacf = go.Figure()
fig_pacf.add_trace(go.Bar(
    x=np.arange(len(lag_pacf)),
    y=lag_pacf,
    marker_color='green'
))

fig_pacf.update_layout(
    title='Partial Autocorrelation Function (PACF)',
    xaxis_title='Lag',
    yaxis_title='PACF',
    template='plotly_white'
)

# Show the plots
fig_acf.show()
fig_pacf.show()
```

**INTERACTIVE TIME SERIES DECOMPOSITION:**Deliveries by different instititions such as private,government etc?

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from statsmodels.tsa.seasonal import seasonal_decompose

# Sample data (replace with actual data)
np.random.seed(42)
dates = pd.date_range('2011-01-01', periods=100, freq='M')
data = {
    'Date': dates,
    'QQ_Institutional_Delivery_Total': np.random.randint(100, 200, size=100),
    'QQ_Delivery_At_Government_Institution_Total': np.random.randint(50, 150, size=100),
    'QQ_Delivery_At_Private_Institution_Total': np.random.randint(30, 80, size=100),
    'QQ_Delivery_At_Home_Total': np.random.randint(20, 60, size=100),
    'QQ_Delivery_At_Home_Conducted_By_Skilled_Health_Personnel_Total': np.random.randint(10, 40, size=100),
    'QQ_Safe_Delivery_Rural': np.random.randint(60, 110, size=100),
    'QQ_Caesarean_Out_Of_Total_Delivery_Taken_Place_In_Government_Institutions_Rural': np.random.randint(10, 50, size=100),
}

df = pd.DataFrame(data)

# Function to perform time series decomposition and plot results
def decompose_and_plot(series, title):
    decomposition = seasonal_decompose(series, model='additive', period=12)

    fig = go.Figure()

    fig.add_trace(go.Scatter(x=series.index, y=decomposition.observed, mode='lines', name='Observed'))
    fig.add_trace(go.Scatter(x=series.index, y=decomposition.trend, mode='lines', name='Trend'))
    fig.add_trace(go.Scatter(x=series.index, y=decomposition.seasonal, mode='lines', name='Seasonal'))
    fig.add_trace(go.Scatter(x=series.index, y=decomposition.resid, mode='lines', name='Residual'))

    fig.update_layout(
        title=title,
        xaxis_title='Date',
        yaxis_title='Value',
        template='plotly_white'
    )

    fig.show()

# Decompose and plot each time series
for column in df.columns[1:]:
    decompose_and_plot(df.set_index('Date')[column], column)

```

**KEY FINDINGS**

Based on the visualizations and analyses conducted using the dataset related to the Indian healthcare system, several key findings can be derived:

**Healthcare Access:**

The majority of deliveries in India occur at home, with a significant proportion being conducted by skilled health personnel.
Institutional deliveries, especially those at government institutions, need to be further promoted to ensure safe delivery practices.

**Child Health:**

There is a prevalence of diarrhoea and acute respiratory infections among children, indicating the need for improved healthcare and preventive measures.
The use of oral rehydration therapy (ORT) and other forms of treatment for diarrhoea needs to be increased, especially in rural areas.

**Maternal Health:**

Antenatal care (ANC) coverage is crucial, and efforts should be made to ensure that pregnant women receive the recommended number of ANC check-ups and tetanus toxoid (TT) injections.
Promotion of iron and folic acid (IFA) consumption among pregnant women is necessary for better maternal and child health outcomes.

**Chronic Illness:**

There is a notable prevalence of chronic illness symptoms among the population, indicating a need for better healthcare services and awareness programs.
Seeking medical care for chronic illnesses, especially among women, needs to be encouraged to improve health outcomes.

**Population Registration:**

Birth registration rates need to be improved, especially in rural areas, to ensure that all children receive their birth certificates.
Family Planning:

There is a need to increase the use of modern contraceptive methods to improve family planning practices and reduce the unmet need for spacing.

**Health Awareness:**

Awareness about HIV/AIDS needs to be increased to reduce its prevalence and improve access to care and support for affected individuals.









**CONCLUSION**

These findings highlight the importance of targeted healthcare interventions and awareness programs to address the healthcare challenges faced by different segments of the population in India. Improving access to healthcare services, promoting preventive measures, and increasing awareness about various health issues are key strategies to improve the overall health and well-being of the population.
